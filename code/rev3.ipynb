```python
import pandas as pd
import numpy as np
import tensorflow as tf
from tensorflow import keras
from matplotlib import pyplot as plt
```

```python
df = pd.read_csv("homeprices_banglore.csv")
df.sample(5)
```

```python
#Important syntax and important transformation
x = df[['area','bedrooms']]
y = df['price']
y = np.array(y)
y = y.reshape(-1,1)
from sklearn import preprocessing

sx = preprocessing.MinMaxScaler()
sy = preprocessing.MinMaxScaler()

scaled_x = sx.fit_transform(x)
scaled_y = sy.fit_transform(y)
y.shape
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(scaled_x,scaled_y,test_size=0.2,random_state=42,shuffle=False)
len(x_test)
print(y_test)
```

```python
model = keras.Sequential([
    keras.layers.Dense(1,input_shape=(2,),activation='linear')
])
model.compile(
    optimizer = 'adam',
    loss = 'mean_squared_error',
    metrics = ['mae']
)
model.fit(x_train,y_train,epochs=20)
```
